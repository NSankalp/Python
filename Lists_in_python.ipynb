```python
fruit=["apple","orange","mango","grapes","banana"]
print(fruit)
print(fruit[0])
print(fruit[-1])
print(fruit[0:3])
print(fruit[:2])
print(fruit[2:])
```

```python
fruit=["apple","orange","mango","grapes","banana"]

(fruit.append("guava")) #adds the fruit at the end
print(type(fruit))
(fruit.insert(3,"kiwi"))#inserts the fruit at the position we want. The number indicates the index value.
print(fruit)
(fruit.remove("guava"))#removes the fruit 
print(fruit)
(fruit.extend("mango"))#divides the string into multiple strings
print(fruit)
(fruit.pop(3))#removes the fruit
print(fruit)
#fruit.clear()#clears the list
#sort
num=[145,20,340,40,510]
num.sort()
print(num)

print(len(fruit))
```

```python
players=["warner","kane","rohit","dhoni"]
print(players)
players.append("conway")
print(players)
players.insert(2,"bumrah")
print(players)
players
```

```python
#sorting

#in ascending order
num=[145,20,340,40,510]
num.sort()
print(num)
#descending order
num.sort(reverse= True)
print(num)
#in ascending order for string
alph=['g','r','k','a']
alph.sort()
print(alph)
#descending order
alph.sort(reverse= True)
print(alph)
```
