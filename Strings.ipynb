```python
a="RRR movie has been released."
print(a[2:5])
print(a[2:])
print(a[:4])
print(a[-2:-5])
```

Modification of strings

```python
a=" RRR movie , has been released. "
print(a.upper())#converts all alphabets into uppercase.
print(a.lower())#coverts all alphabets into lowercase.
print(a.strip())#removes the space after and before the double quotes.
print(a.replace("been",''))
print(a.split(","))
```

Concatenate Strings

```python
a= "David"
b= "Warner"
c= a+b
print(c)
d=a+" "+b
print(d)

```

Format Strings

```python
a=" 18 years "
b= "My name is Sankalp, and I am {}" 
print(b.format(a))

c=185
d=53
e="SRH scored {} runs and won by {} runs"
print(e.format(c,d))

f=185
g=53
h="SRH scored {1} runs and won by {0} runs"
print(h.format(g,f))

```
