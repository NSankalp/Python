## Type Conversions:
### 1)int to float
### 2)float to int
### 3)int to string
### 4)string to int(not possible if its anything other than numbers)


```python
a=10
print(type(a))
b=float(a)
print(b)
print(type(b))

c=23.7
print(type(c))
d=int(c)
print(d)
print(type(d))

x=20
print(type(x))
y=str(x)
print(y)
print(type(y))

s='12'
print(type(s))
t=int(s)
print(t)
print(type(t))


```

## Conversion of boolean
### Boolean to int
### Boolean to float
### Boolean to string

```python
a=10
b=20
c=b<a
d=int(c)
print(c)
print(d)
print(type(c))
print(type(d))
```

```python
a=10
b=20
c=b<a
d=float(c)
print(c)
print(d)
print(type(c))
print(type(d))
```

```python
a=10
b=20
c=b<a
d=str(c)
print(c)
print(d)
print(type(c))
print(type(d))
```

## Complex
### Complex to int(not possible)
### Complex to float(not possible)
### Complex to string(Possible)
### Complex to bool(Possible)

```python
a=1+10j
print(type(a))
```

```python
a=1+10j
print(type(a))
b=str(a)
print(type(b))
```

```python
a=1+10j
print(type(a))
b=bool(a)
print(type(b))
```
